```python
"""
FAQ Knowledge Base Foundation
Task 1 for SupportAI System
"""

def create_faq_knowledge_base():
    """
    Create the FAQ knowledge base with at least 5 entries.

    Returns:
        list: A list of FAQ dictionaries containing id, category,
              question, answer, and keywords fields
    """
    faqs = [
        {
            "id": "faq-001",
            "category": "Account",
            "question": "How do I reset my password?",
            "answer": "Click 'Forgot Password' on the login page. Enter your registered email address and check your inbox for a reset link valid for 24 hours.",
            "keywords": ["password", "reset", "forgot", "login", "credentials", "access"]
        },
        {
            "id": "faq-002",
            "category": "Billing",
            "question": "What is your refund policy?",
            "answer": "We offer full refunds within 30 days of purchase for unused subscriptions. Partial refunds are available for annual plans cancelled after 30 days.",
            "keywords": ["refund", "money back", "return", "cancel", "subscription", "billing"]
        },
        {
            "id": "faq-003",
            "category": "Technical",
            "question": "How can I enable two-factor authentication?",
            "answer": "Go to Settings > Security > Two-Factor Authentication. Enable it using Google Authenticator or SMS verification.",
            "keywords": ["2fa", "two factor", "authentication", "security", "setup", "enable"]
        },
        {
            "id": "faq-004",
            "category": "Billing",
            "question": "How do I upgrade my subscription plan?",
            "answer": "Navigate to Account > Billing > Subscription. Select your desired plan and follow the checkout process. The upgrade takes effect immediately.",
            "keywords": ["upgrade", "plan", "subscription", "billing", "premium", "change"]
        },
        {
            "id": "faq-005",
            "category": "Technical",
            "question": "Why am I getting 'Connection Error' messages?",
            "answer": "Check your internet connection first. If the issue persists, try clearing your browser cache or using a different network. Contact support if the problem continues.",
            "keywords": ["connection", "error", "network", "timeout", "offline", "internet"]
        },
        {
            "id": "faq-006",
            "category": "Account",
            "question": "How can I update my profile information?",
            "answer": "Click on your profile icon in the top right corner, select 'Settings', and update your information in the 'Profile' tab.",
            "keywords": ["profile", "update", "settings", "account", "information", "edit"]
        }
    ]
    return faqs


def search_by_keyword(faqs, query):
    """
    Search FAQs by matching query words against keywords, question, and category fields.
    Matching is case-insensitive. Results are ordered by number of keyword hits (most matches first).

    Args:
        faqs (list): List of FAQ dictionaries
        query (str): Search query string

    Returns:
        list: Matching FAQs ordered by relevance (most keyword hits first)
    """
    # Normalise query to lowercase and split into individual words
    query_words = set(query.lower().split())

    # Store results with their hit counts
    results = []

    for faq in faqs:
        # Combine all searchable text fields
        # Lowercase all fields for case-insensitive comparison
        searchable_text = (
            faq['category'].lower() + " " +
            faq['question'].lower() + " " +
            " ".join([kw.lower() for kw in faq['keywords']])
        )

        # Count how many query words appear in the searchable text
        hit_count = 0
        for word in query_words:
            if word in searchable_text:
                hit_count += 1

        # Only include FAQs that have at least one match
        if hit_count > 0:
            results.append({
                'faq': faq,
                'hit_count': hit_count
            })

    # Sort by hit count in descending order (most hits first)
    results.sort(key=lambda x: x['hit_count'], reverse=True)

    # Return just the FAQ dictionaries
    return [result['faq'] for result in results]


def get_faq_by_id(faqs, faq_id):
    """
    Return the FAQ with the given id, or None if not found.

    Args:
        faqs (list): List of FAQ dictionaries
        faq_id (str): The ID of the FAQ to find

    Returns:
        dict | None: The matching FAQ or None if not found
    """
    for faq in faqs:
        if faq['id'].lower() == faq_id.lower():
            return faq
    return None


def get_faqs_by_category(faqs, category):
    """
    Return all FAQs belonging to the given category (case-insensitive).

    Args:
        faqs (list): List of FAQ dictionaries
        category (str): The category to filter by

    Returns:
        list: All FAQs matching the category
    """
    category_lower = category.lower()
    return [faq for faq in faqs if faq['category'].lower() == category_lower]


def print_search_results(query, results):
    """
    Helper function to display search results in a formatted way.

    Args:
        query (str): The search query that was used
        results (list): List of matching FAQ dictionaries
    """
    print(f"\nQuery: {query}")

    if not results:
        print("  No matching FAQs found.")
    else:
        for faq in results:
            print(f"  [{faq['category']}] {faq['question']}")
            print(f"  → {faq['answer']}")
            print()


def demonstrate_search():
    """
    Demonstrate the search functionality with test queries.
    Tests the search_by_keyword function with various queries.
    """
    # Create the FAQ knowledge base
    faqs = create_faq_knowledge_base()

    print("=" * 60)
    print("FAQ KNOWLEDGE BASE SEARCH DEMONSTRATION")
    print("=" * 60)

    # Test Query 1: Password reset
    results1 = search_by_keyword(faqs, "forgot my password")
    print_search_results("forgot my password", results1)

    # Test Query 2: Refund
    results2 = search_by_keyword(faqs, "refund")
    print_search_results("refund", results2)

    # Test Query 3: No match
    results3 = search_by_keyword(faqs, "weather today")
    print_search_results("weather today", results3)

    # Additional demonstration: Get FAQ by ID
    print("\n" + "=" * 60)
    print("GET FAQ BY ID DEMONSTRATION")
    print("=" * 60)
    faq = get_faq_by_id(faqs, "faq-003")
    if faq:
        print(f"Found FAQ: [{faq['category']}] {faq['question']}")
        print(f"Answer: {faq['answer']}")

    # Additional demonstration: Get FAQs by category
    print("\n" + "=" * 60)
    print("GET FAQS BY CATEGORY DEMONSTRATION")
    print("=" * 60)
    billing_faqs = get_faqs_by_category(faqs, "Billing")
    print(f"Found {len(billing_faqs)} FAQs in the Billing category:")
    for faq in billing_faqs:
        print(f"  - {faq['question']}")


if __name__ == "__main__":
    demonstrate_search()
```
